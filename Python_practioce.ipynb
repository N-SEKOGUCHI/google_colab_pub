```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 2 * x  # xに2をかけてy座標とする

x = np.linspace(-5, 5)  # -5から5まで
y = f(x)

plt.plot(x, y, label='$y = 2*x$')
plt.legend()
plt.show()
```
