```python
## My first Python　with NumPy and matplotlib

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 2 * x**2  # xに2をかけてy座標とする

x = np.linspace(-5, 5)  # -5から5まで
y = f(x)

# Assert that y is not empty
assert y.size > 0, "The array y should not be empty."

plt.plot(x, y, label='$ y = 2x^2 $')
plt.legend()
plt.show()
```

# 連立一次方程式

### 解が一意に決まる
$ \qquad
\left\{\begin{array}{l}
 y = -\frac{1}{3}x + \frac{5}{3} \\
 y = \frac{2}{3}x - \frac{1}{3}
\end{array}\right.
 $

## 解が無限にある
$ \qquad
\left\{\begin{array}{l}
 y = -\frac{1}{3}x + \frac{5}{3} \\
 y = -\frac{1}{3}x + \frac{5}{4}
\end{array}\right.
 $

## 解が無い
$ \qquad
\left\{\begin{array}{l}
 y = -\frac{1}{3}x + \frac{5}{3} \\
 y = -\frac{1}{3}x + \frac{5}{4}
\end{array}\right.
 $


```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
  return -(1/3) * x + (5/3)

def f2(x):
  return (2/3) * x - (1/3)

x = np.linspace(-5, 5)  # -5から5まで
y1 = f1(x)
y2 = f2(x)

plt.plot(x,y1,color='blue',label='$ y = -\\frac{1}{3}x + \\frac{5}{3} $')
plt.plot(x,y2,color='red',label='$ y = \\frac{2}{3}x - \\frac{1}{3} $')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
  return -(1/3) * x + (5/3)

def f2(x):
  return (- x + 5)/3

x = np.linspace(-5, 5)  # -5から5まで
y1 = f1(x)
y2 = f2(x)

plt.plot(x,y1,color='blue',label='$ y = -\\frac{1}{3}x + \\frac{5}{3} $')
plt.plot(x,y2,color='red',linestyle='dotted',linewidth=3,label='$y=\\frac{-x+5}{3}$')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
  return -(1/3) * x + (5/3)

def f2(x):
  return -(1/3) * x + (5/4)

x = np.linspace(-5, 5)  # -5から5まで
y1 = f1(x)
y2 = f2(x)

plt.plot(x,y1,color='blue',label='$ y = -\\frac{1}{3}x + \\frac{5}{3} $')
plt.plot(x,y2,color='red',label='$ y = -\\frac{1}{3}x + \\frac{5}{4} $')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
