```python
## My first Python　with NumPy and matplotlib

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 2 * x**2  # xに2をかけてy座標とする

x = np.linspace(-5, 5)  # -5から5まで
y = f(x)

# Assert that y is not empty
assert y.size > 0, "The array y should not be empty."

plt.plot(x, y, label='$ y = 2x^2 $')
plt.legend()
plt.show()
```
